```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import json
import openpyxl
import re

option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs", prefs)


driver = webdriver.Chrome("C:\\Users\\Ting-Wei Yu\\chromedriver.exe", options = option)
driver.get("https://www.ubereats.com/tw")


picSearchBar = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'input[name="searchTerm"]')))
picSearchBar.send_keys("台北市文山區")

time.sleep(1)
picSearchBar.send_keys('\ue007')     # 按下Enter

time.sleep(3)


# 點下熱門餐點radio button，為了先進一步排除前幾類自動推薦的餐廳(排列順序太雜亂不好爬)
hot = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[text()="  評分"]')))
hot.click()

time.sleep(3)

# 顯示更多餐廳到不能顯示
# 測試時先限制迴圈次數，while限制迴圈次數就用個 count 計數器


while True: 
    try: 
        more = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[text()="顯示更多餐廳"]')))
        more.click()
        time.sleep(6)
    except: 
        break
        
```

```python
# 找出所有餐廳頁面的網址 + 存成excel

wb = openpyxl.Workbook()
ws = wb.active

stores = driver.find_elements(By.CSS_SELECTOR, 'a[data-testid="store-card"]')

count = 0

ws["A1"] = "URL"

href = []

for i in stores: 
    href.append(i.get_attribute("href"))
    storeURL = i.get_attribute("href")
    
    time.sleep(1)
    count += 1
    
    print("===== storeURL", count, "=====")
    
    ws.append([storeURL])
    wb.save("Go_Taipei文山區_noon_storeURL.xlsx")

    
print(href)
```
