```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import json
import openpyxl
import re
import pandas as pd

option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs", prefs)

def data_clean(text):
    # 清洗excel中的非法字符，都是不常见的不可显示字符，例如退格，响铃等
    ILLEGAL_CHARACTERS_RE = re.compile(r"[\000-\010]|[\013-\014]|[\016-\037]")
    text = ILLEGAL_CHARACTERS_RE.sub(r"", text)
    return text

driver = webdriver.Chrome("C:\\Users\\Ting-Wei Yu\\chromedriver.exe", options = option)
driver.get("https://www.ubereats.com/tw")


wb = openpyxl.Workbook()
ws = wb.active

ws["A1"] = "餐廳名稱"
ws["B1"] = "餐廳類型"
ws["C1"] = "餐廳總評分"
ws["D1"] = "地址"
ws["E1"] = "經度"
ws["F1"] = "緯度"
ws["G1"] = "訂餐網址"


df = pd.read_excel("Go_文山信義南港_noon_storeURL.xlsx")
urls = df["URL"]      # 提取網址欄位的數據


count = 0

for i in urls[1343:]: 
    driver.get(i)
    time.sleep(3)
    
    try:     
        detail  = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[text()="詳細資訊"]')))
        detail.click()

        time.sleep(5)

        soup = BeautifulSoup(driver.page_source, "html.parser")       # 以 "html.parser" 的狀態開啟 driver.page_source
        # print(soup)

        info = soup.find_all("main", id="main-content")[0].script.text
        dicInfo = json.loads(info)

        storeName = dicInfo["name"]                           # 店名
        
        try: 
            foodType = dicInfo["servesCuisine"][0] + " / " + str(dicInfo["servesCuisine"])                # 類型
        except: 
            foodType = ""      
        
        try: 
            longitude = dicInfo["geo"]["longitude"]               # 經度
        except: 
            longitude = ""              
        
        try: 
            latitude = dicInfo["geo"]["latitude"]                 # 緯度
        except: 
            latitude = ""              
        
        try: 
            rating = dicInfo["aggregateRating"]["ratingValue"]    # 總評分
        except: 
            rating = ""    # 若抓不到總評分（會error），則設定總評分為空值
            
        count += 1
        print("===== STORE", count, "=====")

        time.sleep(1)
        
    except: 
        closeButton = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[aria-label="關閉"]')))
        closeButton.click()
        
        time.sleep(2)
        
        detail  = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[text()="詳細資訊"]')))
        detail.click()

        time.sleep(5)

        soup = BeautifulSoup(driver.page_source, "html.parser")       # 以 "html.parser" 的狀態開啟 driver.page_source
        # print(soup)

        info = soup.find_all("main", id="main-content")[0].script.text
        dicInfo = json.loads(info)

        storeName = dicInfo["name"]                           # 店名
        
        try: 
            foodType = dicInfo["servesCuisine"][0] + " / " + str(dicInfo["servesCuisine"])                # 類型
        except: 
            foodType = ""      
        
        try: 
            longitude = dicInfo["geo"]["longitude"]               # 經度
        except: 
            longitude = ""              
        
        try: 
            latitude = dicInfo["geo"]["latitude"]                 # 緯度
        except: 
            latitude = ""              
        
        try: 
            rating = dicInfo["aggregateRating"]["ratingValue"]    # 總評分
        except: 
            rating = ""    # 若抓不到總評分（會error），則設定總評分為空值 
            
        count += 1
        print("===== STORE", count, "=====")

        time.sleep(1)
        
        
    ws.append([data_clean(storeName), foodType, rating, "", longitude, latitude, i])
    
    wb.save("Go_文山信義南港_noon_part3.xlsx")
```
